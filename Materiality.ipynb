```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# ---------------- GRID (15 scale) ----------------
x = np.linspace(1, 5, 200)
y = np.linspace(1, 5, 200)
X, Y = np.meshgrid(x, y)

# Combined materiality score
Z = X + Y

# ---------------- COLOR MAP (brown gradient) ----------------
cmap = ListedColormap([
    "#f3ede7",  # Low (very light brown)
    "#d6bfae",  # Medium (warm brown)
    "#8b5a2b"   # High (dark brown)
])

# Thresholds for Low / Medium / High
levels = [0, 5.5, 7.5, 11]

# ---------------- FIGURE ----------------
fig, ax = plt.subplots(figsize=(8, 6))

# Background materiality zones
contour = ax.contourf(
    X, Y, Z,
    levels=levels,
    cmap=cmap,
    alpha=0.9
)

# ---------------- DATA POINTS ----------------
topics = [
    "Circularity & waste",
    "Biodiversity & land use",
    "Community impact",
    "Own workforce wellbeing",
    "DEI (diversity & inclusion)",
    "Water & chemicals",
    "Product safety",
    "Business ethics",
    "Data privacy",
    "Wages (supply chain)",
    "H&S (supply chain)",
    "Climate & GHG",
]

x_vals = [2.1, 1.8, 3.4, 2.6, 1.9, 3.2, 4.1, 3.3, 4.4, 3.6, 4.3, 4.7]
y_vals = [2.3, 3.6, 4.5, 2.4, 3.2, 4.3, 2.1, 3.4, 4.2, 3.5, 4.4, 4.8]

colors = [
    "navy","navy","blue",
    "blue","blue","navy",
    "blue","magenta","magenta",
    "blue","blue","navy"
]

ax.scatter(
    x_vals, y_vals,
    c=colors,
    s=120,
    edgecolor="white",
    zorder=3
)

for x0, y0, label in zip(x_vals, y_vals, topics):
    ax.text(x0 + 0.05, y0 + 0.05, label, fontsize=8)

# ---------------- AXES ----------------
ax.set_xlim(1, 5)
ax.set_ylim(1, 5)
ax.set_xticks([1,2,3,4,5])
ax.set_yticks([1,2,3,4,5])

ax.set_xlabel("Financial Materiality")
ax.set_ylabel("Impact Materiality")
ax.set_title("Double Materiality Matrix")

# ---------------- ZONE LABELS ----------------
ax.text(1.2, 1.2, "LOW", fontsize=14, weight="bold", color="#6b4a2d")
ax.text(2.6, 2.6, "MEDIUM", fontsize=14, weight="bold", color="#6b4a2d")
ax.text(3.9, 3.9, "HIGH", fontsize=14, weight="bold", color="white")

plt.tight_layout()
plt.show()
```
