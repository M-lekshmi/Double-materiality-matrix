```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D

# ---------------- GRID (15 scale) ----------------
x = np.linspace(1, 5, 200)
y = np.linspace(1, 5, 200)
X, Y = np.meshgrid(x, y)

Z = X + Y

# ---------------- COLOR MAP ----------------
cmap = ListedColormap([
    "#f7f3ef",  # Low
    "#e4d3c5",  # Medium
    "#9a6a3a"   # High
])

levels = [0, 5.5, 7.5, 11]

# ---------------- FIGURE ----------------
fig, ax = plt.subplots(figsize=(8, 6))

ax.contourf(
    X, Y, Z,
    levels=levels,
    cmap=cmap,
    alpha=0.25,
    linewidths=0,
    antialiased=False
)

# ---------------- DATA ----------------
topics = [
    "Circularity & waste",
    "Biodiversity & land use",
    "Community impact",
    "Own workforce wellbeing",
    "DEI (diversity & inclusion)",
    "Water & chemicals",
    "Product safety",
    "Business ethics",
    "Data privacy",
    "Wages (supply chain)",
    "H&S (supply chain)",
    "Climate & GHG",
]

# Adjusted coordinates (Community impact nudged)
x_vals = [
    2.8,  # Circularity & waste
    2.3,  # Biodiversity & land use
    3.0,  # Community impact
    2.4,  # Own workforce wellbeing
    2.0,  # DEI
    3.3,  # Water & chemicals
    4.1,  # Product safety
    3.4,  # Business ethics
    4.4,  # Data privacy
    3.7,  # Wages (supply chain)
    4.3,  # H&S (supply chain)
    4.7,  # Climate & GHG
]

y_vals = [
    2.6,  # Circularity & waste
    3.8,  # Biodiversity & land use
    4.55, # Community impact
    2.8,  # Own workforce wellbeing
    3.2,  # DEI
    4.4,  # Water & chemicals
    3.0,  # Product safety
    3.6,  # Business ethics
    4.3,  # Data privacy
    3.7,  # Wages (supply chain)
    4.5,  # H&S (supply chain)
    4.8,  # Climate & GHG
]

colors = [
    "navy",     # Environment
    "navy",     # Environment
    "blue",     # Social
    "blue",     # Social
    "blue",     # Social
    "navy",     # Environment
    "blue",     # Social
    "magenta",  # Governance
    "magenta",  # Governance
    "blue",     # Social
    "blue",     # Social
    "navy"      # Environment
]

ax.scatter(
    x_vals, y_vals,
    c=colors,
    s=120,
    edgecolor="white",
    zorder=3
)

# Labels
for x0, y0, label in zip(x_vals, y_vals, topics):
    ax.text(x0 + 0.05, y0 + 0.05, label, fontsize=8)

# ---------------- AXES ----------------
ax.set_xlim(1, 5)
ax.set_ylim(1, 5)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_yticks([1, 2, 3, 4, 5])

ax.set_xlabel("Financial Materiality")
ax.set_ylabel("Impact Materiality")
ax.set_title("Double Materiality Matrix (Aligned with H&M Group 2024)")

# ---------------- ZONE LABELS ----------------
ax.text(1.2, 1.2, "LOW", fontsize=14, weight="bold", color="#6b4a2d")
ax.text(2.6, 2.6, "MEDIUM", fontsize=14, weight="bold", color="#6b4a2d")
ax.text(3.9, 3.9, "HIGH", fontsize=14, weight="bold", color="white")

# ---------------- ESG LEGEND ----------------
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Environment',
           markerfacecolor='navy', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Social',
           markerfacecolor='blue', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Governance',
           markerfacecolor='magenta', markersize=10),
]

ax.legend(
    handles=legend_elements,
    title="ESG Category",
    loc="lower right",
    frameon=False
)

plt.tight_layout()
plt.show()
```
